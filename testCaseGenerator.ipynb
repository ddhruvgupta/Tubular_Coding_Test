```python
import numpy as np
import json
```

```python
k = 65 + np.random.random ([50,50]) * (90-65)
k = k.astype(int)
new_k = k.tolist()

for i in range(0,len(new_k)):
    for j in range(0,len(new_k[i])):
            new_k[i][j] = chr(new_k[i][j])
```

```python
new_k
```

```python
# import pickle

with open('outfile.txt', 'w') as fp:
    json.dump(new_k,fp)
```
